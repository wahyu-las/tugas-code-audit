```python
# ini program buat hitung harga
def hitung():
    # harga komponen
    x = 500000  # harga ram
    y = 1500000 # harga prosesor
    z = 2000000 # harga vga

    # diskon 10% buat ram
    d1 = x - (x * 0.1)

    # diskon 10% buat vga
    d2 = z - (z * 0.1)

    # hitung total
    t = d1 + y + d2 + 150000 # 150000 itu pajak

    # ini buat ngeprint hasilnya
    print("Total harga rakit PC: Rp", t)

    if t > 3500000:
        print("Kategori: Mahal")
    else:
        print("Kategori: Murah")

hitung()
```
