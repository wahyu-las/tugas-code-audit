```python
# Konstanta untuk menghindari Magic Numbers
DISCOUNT_RATE = 0.10
TAX_FEE = 150000
EXPENSIVE_THRESHOLD = 3500000

def calculate_discounted_price(price):
    """Fungsi khusus untuk menghitung diskon (DRY Principle)"""
    return price - (price * DISCOUNT_RATE)

def print_price_category(total_price):
    """Fungsi khusus untuk mencetak kategori (SRP Principle)"""
    if total_price > EXPENSIVE_THRESHOLD:
        print("Kategori: Mahal")
    else:
        print("Kategori: Murah")

def calculate_total_pc_price():
    # Menggunakan Meaningful Names
    ram_price = 500000
    cpu_price = 1500000
    vga_price = 2000000

    # Memanggil fungsi agar tidak ada kode berulang
    discounted_ram = calculate_discounted_price(ram_price)
    discounted_vga = calculate_discounted_price(vga_price)

    total_price = discounted_ram + cpu_price + discounted_vga + TAX_FEE

    # Output yang bersih
    print(f"Total harga rakit PC: Rp {int(total_price)}")
    print_price_category(total_price)

# Menjalankan program
if __name__ == "__main__":
    calculate_total_pc_price()
```
